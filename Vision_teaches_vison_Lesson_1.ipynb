```python
#importing required modules
import numpy as np
import urllib
import cv2
from matplotlib import pyplot as plt
```

```python
#converting url to an image usable in colab
def url_to_image(url):
    resp = urllib.request.urlopen(url)
    image = np.asarray(bytearray(resp.read()), dtype="uint8")
    image = cv2.imdecode(image, cv2.IMREAD_COLOR)
    corrected_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
 
    return corrected_image
```

```python
#url of image 
#try with your own images (you can use imgur to host your images)
url = "https://i.imgur.com/JtCUHTM.jpg"
#conversion of image
image = url_to_image(url)
#displaying image
plt.imshow(image)
plt.show()
```

```python
#An image is just a collection of pixels and we can change values of these pixels manually
corner = image[0:100, 0:100]
plt.imshow(corner)
plt.show()
```

```python
#here look closely the soul stone is gone..!
copy_image = image.copy()
copy_image[30:80, 280:340] = (121, 21, 43)
plt.imshow(copy_image)
plt.show()
```

```python
#You can also draw on images this step is crucial for visual cue for detection of an object
#eg. Face Detection
copy_image = image.copy()
rectangle = cv2.rectangle(copy_image,(200,130),(560,450),(0,255,0),3)
plt.imshow(rectangle)
plt.show()
```
